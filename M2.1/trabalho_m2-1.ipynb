```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Link para Dataset: https://www.kaggle.com/datasets/thedevastator/mcdonalds-ice-cream-machines-broken-timeseries

from pandas import DataFrame


maquinas_mcdonalds = pd.read_csv("./mcdonalds_dataset.csv")

paises = np.sort(maquinas_mcdonalds["country"].unique())

total_maquinas = maquinas_mcdonalds["country"].count()
maquinas_por_pais = maquinas_mcdonalds.sort_values("country").groupby("country")
plt.title("Total de máquinas")
plt.bar(paises, maquinas_por_pais["city"].count())
plt.show()

filtro_maquinas_quebradas = maquinas_mcdonalds["is_broken"] == True
maquinas_quebradas = maquinas_mcdonalds[filtro_maquinas_quebradas]
groupby = maquinas_mcdonalds[filtro_maquinas_quebradas].sort_values("country").groupby("country")
maquinas_quebradas.to_csv("./maquinas_quebradas.csv", index=False)
plt.title("Contagem de 4 países com máquinas quebradas")
plt.bar(paises, groupby["city"].count())
plt.show()

filtro_maquinas_funcionando = maquinas_mcdonalds["dot"] == "working"
filtro_maquinas_usa = maquinas_mcdonalds["country"] == "USA"
maquinas_funcionado = maquinas_mcdonalds[filtro_maquinas_funcionando]
maquinas_funcionado_usa = maquinas_mcdonalds[filtro_maquinas_usa]
maquinas_funcionado_usa.to_csv("./maquinas_funcionando_usa.csv", index=False)
groupby = maquinas_mcdonalds[filtro_maquinas_funcionando].groupby("state")
contagem_maquinas_funcionando = groupby["dot"].count()
plt.title("Contagem de 20 estados no USA com máquinas funcionando")
plt.plot(contagem_maquinas_funcionando[:20])
plt.show()
```
